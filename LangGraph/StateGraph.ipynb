```python
from langgraph.graph import START, END, StateGraph
from typing_extensions import TypedDict
from langchain_openai.chat_models import ChatOpenAI
from langchain_core.messages import HumanMessage, BaseMessage
from langchain_core.runnables import Runnable
from collections.abc import Sequence
```

```python
# defining the state

class State(TypedDict):
    messages : Sequence[BaseMessage]

state = State(messages = [HumanMessage("Tell me a joke!")])    
```

```python
# defining the openai chat api

chat = ChatOpenAI(
    model = "gpt-4o",
    seed = 365,
    temperature= 0,
    max_completion_tokens=100
)
```

```python
# defining the node

def chatbot(state: state) -> State:
    response = chat.invoke(state['messages'])

    response.pretty_print()

    return State(messages = [response])
```

# Defining the Graph

```python
graph = StateGraph(State)
```

```python
graph.add_node("chatbot", chatbot)

graph.add_edge(START, "chatbot")

graph.add_edge("chatbot", END)

graph_compiled = graph.compile()
```

```python
isinstance(graph_compiled, Runnable)

# isinstance(graph, Runnable) -> will return false as the graph is just a builder and not yet compiled and hence not a runnable.
```

```python
graph_compiled.invoke(state)
```
