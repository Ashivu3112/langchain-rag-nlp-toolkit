```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import spacy
import re
import pandas as pd
import matplotlib.pyplot as plt
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
raw_df = df
raw_df.head()
```

```python
raw_filter = raw_df.filter(['title'])
raw_filter.head()
```

# Creating tokens from the text

```python
tokenised_df = raw_filter.copy()
tokenised_df['tokens'] = tokenised_df['title'].map(lambda x: word_tokenize(x))
tokenised_df.head()
```

# Removing stop words

```python
len(stopwords.words('english'))
```

# Iterate through the tokens, convert them all to lower case, remove the stop words and join them.

```python
tokenised_df['no_stop_words'] = tokenised_df['tokens'].map(lambda x: [word for word in x if word.lower() not in stopwords.words('english')])
tokenised_df['text'] = tokenised_df['no_stop_words'].map(lambda x: ' '.join(x))
tokenised_df['text_no_punc'] = tokenised_df.apply(lambda x: re.sub(r"([^\w\s])", "", x['text']), axis=1)

#reorder the columns
tokenised_df = tokenised_df[['title', 'tokens', 'no_stop_words', 'text', 'text_no_punc']]

tokenised_df.head()
```

# POS and NER Tagging

```python
nlp = spacy.load("en_core_web_sm")
```

```python
text = " ".join(tokenised_df['text_no_punc'])
text
```

```python
doc = nlp(text)
```

POS

```python
pos_df = pd.DataFrame(columns = ['token', 'pos'])

for token in doc:
  if token.text.strip():
    pos_df = pd.concat([pos_df, pd.DataFrame.from_records([{'token': token.text, 'pos': token.pos_}])], ignore_index=True)

pos_df.head()
```

```python
# token frequency count
pos_df_counts = pos_df.groupby(['token','pos']).size().reset_index(name='counts').sort_values(by='counts', ascending=False)
pos_df_counts.head(10)
```

NER

```python

# extract the tokens and entity tags into a dataframe
ner_df = pd.DataFrame(columns=['token', 'ner_tag'])

for token in doc.ents:
    if pd.isna(token.label_) is False:
        ner_df = pd.concat([ner_df, pd.DataFrame.from_records(
            [{'token': token.text, 'ner_tag': token.label_}])], ignore_index=True)

ner_df.head(10)
```
